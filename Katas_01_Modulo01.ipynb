# Ejercicio número 2: Convertidor de unidades
La Superior requiere una calculadora que sea capaz de convertir los años luz en parsec.

```python
print("Parsec - Light-years conversor")
first_number = input("How many Parsec do you want convert? ")
second_number = float (first_number) * 3.26156
print(str(first_number) + " parsec, is " + str(second_number) + " light-years")
```
