Operadores aritméticos en Python

Python proporciona operadores aritméticos comunes para que puedas realizar operaciones matemáticas en tu código. Estos incluyen las cuatro operaciones principales de suma, resta, multiplicación y división.

Exploremos cómo podemos crear un programa que pueda calcular la distancia entre dos planetas. Comenzaremos usando dos distancias de planetas: Tierra (149.597.870 km) y Júpiter (778.547.200 km).

```python
tierra = 149597870
jupiter = 778547200

distancia = jupiter - tierra
print(distancia, "km")
```

```python
# Distancia en millas
millas = distancia * 0.621
print(millas, "millas")
```

# Ejercicio 2: convierte cadenas en números y usa valores absolutos

```python
planeta_1 = input("Introduzca nombre del planeta:")
distancia_1 = int(input(f'Introduzca distancia del sol al planeta {planeta_1}:'))
planeta_2 = input("Introduzca nombre del segundo planeta:")
distancia_2 = int(input(f'Introduzca distancia del sol al planeta {planeta_2}:'))

distancia = (abs(distancia_1 - distancia_2)) * 0.621

print(f'La distanica entre {planeta_1} y {planeta_2} es de {distancia} millas.')


```
