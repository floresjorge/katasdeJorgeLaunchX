# Ejercicio1: Crear y usar listas de Python




En primer lugar, crea una variable denominada planets. Agrega los ocho planetas (sin Plutón) a la lista. A continuación, muestra el número de planetas.


```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
```

```python
# Número de planetas

no_planets = len(planets)
print(f'There are {no_planets} planets on the Solar System.')
```

```python
# Muestra número de planetas y el último
planets.append('Pluto')
no_planets = len(planets)
print(f'There are {no_planets} planets on the Solar System. The last one is {planets[9]}')
```

# Ejercicio 2: Trabajando con datos de una lista

```python
#Lista de planetas
planetas = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano' 'Neptuno']
```

```python
# Solicitar el ingreso de un planeta
planeta = input("Por favor, ingresa el nombre, en español, del planeta:").title()
print(planeta)
```

```python
#Buscar el planeta en la lista
planeta_index = planetas.index(planeta)
print(f' {planeta} está en la posición número {planeta_index} con respecto al Sol.')
```

```python
#Mostrar los planetas más cercanos al Sol
print('Los planetas más cercanos que ' + planeta + ' son:')
print(planets[0:planeta_index])
```

```python
#Mostrar los planeras más lejanos al Sol
print('Los planetas más alejados que ' + planeta + ' son:')
print(planets[planeta_index + 1:])
```
